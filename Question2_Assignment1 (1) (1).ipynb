```python
import pandas as pd
df =pd.read_csv('house_data.csv')
df

```

```python
df.drop(columns=['ID'],inplace=True)
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
# df.columns
num_col = ['Square_Feet','Garage_Size','Location_Score','Distance_to_Center']
categ_col = ['Num_Bedrooms','Num_Bathrooms','Num_Floors','Year_Built','Has_Garden','Has_Pool']
```

```python
for num in num_col:
    df[num].fillna(np.mean(df[num]),inplace=True)
for cat in categ_col:
    df[cat] = df[cat].fillna(df[cat].mode()[0])
```

```python
cols = num_col
for col in cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(df[col])
```

```python
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0, annot=True, fmt='.2f')
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df)
```

We can see Square_feet, Num_Bedrooms and Year_Built have positive correlation with Price.

```python
def normalize(col):
    col_mean = col.mean()
    col_std = col.std()
    col = (col-col_mean)/col_std
    return col
```

```python
for c in num_col:
    df[c] = normalize(df[c])
    # print(c)
```

```python
df
```

```python
# X = df[['Square_Feet','Num_Bedrooms','Year_Built']]
X = df.drop(columns=['Price'])
X
```

```python
Y = df['Price']
```

```python
# Add intercept term (column of ones)
X_np = np.c_[np.ones(X.shape[0]), X.to_numpy()]  # Shape: (n_samples, n_features+1)

# Convert y to numpy array
Y_np = Y.to_numpy().reshape(-1, 1)  # Shape: (n_samples, 1)

# Normal Equation: theta = (X^T X)^(-1) X^T y
theta = np.linalg.inv(X_np.T @ X_np) @ X_np.T @ Y_np

print("Theta (coefficients):")
print(theta)
```

```python
# Predictions
Y_pred = X_np @ theta
```

Interpreting model's performance

```python
# Evaluation (Mean Squared Error)
mse = np.mean((Y_np - Y_pred)**2)
print(f"Mean Squared Error: {mse:.2f}")

# Calculating Root Mean Square
rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)

# Calculating Residual Sum of Squares
n_samples = X.shape[0]
rss = mse * n_samples
print("Residual Sum of Squares (RSS):", rss)
```

Plotting Predicting Prices VS Actual Prices

```python
plt.figure(figsize=(6, 4))
plt.scatter(Y_np, Y_pred, alpha=0.5)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Housing Prices")
plt.plot([Y_np.min(), Y_np.max()], [Y_np.min(), Y_np.max()], color='red')
plt.grid(True)
plt.show()
```
